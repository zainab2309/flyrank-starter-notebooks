## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

The starter dataset contains 30,000 content pages, 44 columns, and 32 clients. The dataset includes pages with different performance trends. The average CTR is 51%, and the average number of impressions over the last 90 days is 5200.37. These differences suggest there is enough variation in page performance to build a ranking system that helps editors prioritize which pages to review first.

```python
import pandas as pd

url = "https://raw.githubusercontent.com/zainab2309/flyrank-starter-notebooks/main/data/raw/content_refresh_anonymized.csv"

df = pd.read_csv(url)

print("Rows:", len(df))
print("Columns:", len(df.columns))
print("Clients:", df["client_id"].nunique())

print("\nTrend direction:")
print(df["trend_direction"].value_counts())

print("\nAverage CTR (%):")
print(round(df["ctr"].mean(), 2))

print("\nAverage impressions (90d):")
print(round(df["impressions_90d"].mean(), 2))
```

## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

My project is a ranking/scoring task. The goal is to rank content pages by priority so that SEO specialists and content editors know which pages to review first. A ranked list is more useful than a simple yes/no prediction because editors usually have limited time and need to decide where to focus their efforts.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd

url = "https://raw.githubusercontent.com/zainab2309/flyrank-starter-notebooks/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(url)

print("Dataset shape:", df.shape)
print("Unique content pages:", df["content_id"].nunique())
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

The model will produce a priority score rather than a class label. For this starter dataset, the ranking is based on observed performance indicators such as impressions, clicks, CTR, and engagement metrics. The trend_direction and trend_pct columns describe page performance, but they are derived values and should not be used as input features when training a model.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
print(df[[
    "impressions_90d",
    "clicks_90d",
    "ctr",
    "engagement_rate"
]].describe())
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

I will evaluate the ranking using Precision@K. A good model should place genuinely high-priority pages near the top of the ranked list so editors spend their time on the most valuable pages first.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
print("Total pages:", len(df))
print("Average CTR (%):", round(df["ctr"].mean(), 2))
print("Average Position:", round(df["avg_position"].replace(0, pd.NA).mean(), 2))
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A simple rule such as "refresh pages with low CTR" would ignore many other useful signals. Page performance depends on multiple factors including impressions, average position, content age, word count, engagement, and search demand. These relationships interact in complex ways, making a machine learning model better suited for ranking pages than a single if-statement.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
features = [
    "search_volume",
    "impressions_90d",
    "clicks_90d",
    "ctr",
    "avg_position",
    "engagement_rate",
    "content_age_days"
]

print(df[features].describe())

```
